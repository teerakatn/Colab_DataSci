```python
import pandas as pd
w_restaurant=pd.read_csv('w_restaurant.csv')
```

```python
w_restaurant.head(5)#ดูdataบางส่วน
```

```python
#ดูขนาด data ทั้งหมด ขนาด
print(w_restaurant.shape)
```

```python
#การเลือก colum
select_column=['id','name','weighted_average_rating','category_name']
w_restaurant_selection=w_restaurant.loc[:,select_column]
w_restaurant_selection.head(5)
```

```python
print(w_restaurant.columns) #ดูชื่อcolumnทั้งหมด
```

```python
colum=['id','name','number_of_reviews','category_name','number_of_checkins']
row=w_restaurant.loc[:,colum] #การเลือกแถว และคอลัมน์
row.head(10)
```

```python
#เลือกเเถวจากเงื่อนไขที่เป็นตัวเลข ต้องการเฉพาะเเถวที่มี weighted_average_rating>3.0
condition=w_restaurant['weighted_average_rating']>3.0
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#เลือกเฉพาะจำนวนที่เกิน10
condition=w_restaurant['number_of_reviews']>10
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#เงื่อนไขหาคอลัมข้อมูลประเภทข้อความ
#หาคอลัมที่ขึ้นต้นด้วยคำว่าก๋วยเตี๋ยว
condition=w_restaurant['name'].str.startswith('ก๋วยเตี๋ยว')
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#หาร้านที่อยู่ในจังหวัดที่มีคำว่านครนำหน้า startswith
condition=w_restaurant['city_name'].str.startswith('Nakhon')
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#ตัวอย่าง name ลงท้ายด้วยคำว่า ดงมูลเหล็ก endswith
condition=w_restaurant['name'].str.endswith('ดงมูลเหล็ก')
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#ตัวอย่าง name มีคำว่าขาหมูอยู่ในข้อความ contains
condition=w_restaurant['name'].str.contains('ขาหมู')
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#zipcode หาร้านที่เป็นค่าว่าง isna()
condition=w_restaurant['zipcode'].isna()
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#เชื่อมเงื่อนไข and or not
#หาร้านที่มีเงื่อนไข2อย่าง 1.zipcode ที่ค่าไม่ว่าง notna  2.60อยุ่ในzipcode ใช้ contai
condition1=w_restaurant['zipcode'].notna()
condition2=w_restaurant['zipcode'].str.contains('60')
condition=condition1 & condition2
w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#เชื่อมเงื่อนไข and or not
#หาร้านที่มีเงื่อนไข2อย่างใน weighted_average_ratin  1.rating ระหว่าง2-4
condition=(w_restaurant['weighted_average_rating']>=2.0) & (w_restaurant['weighted_average_rating']<=4.0)

select_column=['id','name','weighted_average_rating']
w_restaurant_selection=w_restaurant.loc[condition, select_column]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#หาร้านที่มีคำว่าขาหมูหรือก๋วยเตี๋ยว
condition=(w_restaurant['name'].str.contains('ขาหมู')) | (w_restaurant['name'].str.contains('ก๋วยเตี๋ยว'))

w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(5)
```

```python
#หาร้านที่อยู่ในกรุงเทพมีเรตติ้ง>4 หรือ อยู่เชียงใหม่เรตติ้ง>3
condition1=(w_restaurant['city_name']=='Bangkok Metropolitan Region') & (w_restaurant['weighted_average_rating']>4.0)
condition2=(w_restaurant['city_name'].str.contains('Chiang Mai')) & (w_restaurant['weighted_average_rating']>3.0)
condition=condition1 | condition2

w_restaurant_selection=w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```
